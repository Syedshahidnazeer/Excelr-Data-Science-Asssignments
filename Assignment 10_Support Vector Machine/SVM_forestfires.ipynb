```python
x = data.iloc[:,2:11]
y = data['size_category']
```

```python
from sklearn.preprocessing import StandardScaler
stdscaler_model = StandardScaler().fit_transform(x)
```

```python
df_x=pd.DataFrame(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(stdscaler_model,y,test_size=0.2,random_state=5)
```

```python
from sklearn.svm import SVC
linear_model =SVC(kernel='linear')
poly_model = SVC(kernel='poly')
rbf_model = SVC(kernel='rbf')
```

```python
linear_model.fit(x_train,y_train)
poly_model.fit(x_train,y_train)
rbf_model.fit(x_train,y_train)
```

```python
Linear_score=linear_model.score(x_test,y_test)
Poly_score=poly_model.score(x_test,y_test)
rbf_score=rbf_model.score(x_test,y_test)
```

```python
#make a table containing scores for each prepared model.
scores={'SVM Kernel':['Linear','Poly','rbf'],'scores':[Linear_score,Poly_score,rbf_score]}
table=pd.DataFrame(scores)
table

# as we can see in the table that the linear kernal with svm has the highest accuracy scores.
```
