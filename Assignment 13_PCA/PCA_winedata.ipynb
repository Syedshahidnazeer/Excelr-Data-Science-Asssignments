```python
# importing all the packages and naming them as per our convinience.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
"""Perform Principal component analysis and perform clustering using first 
3 principal component scores (both heirarchial and k mean clustering(scree plot or elbow curve) and obtain 
optimum number of clusters and check whether we have obtained same number of clusters with the original data 
(class column we have ignored at the begining who shows it has 3 clusters)df"""
```

```python
# Load the dataset using pandas
data=pd.read_csv('E:\\PCA\\wine.csv')
data
```

```python
# creating a dataframe using pandas.
df=pd.DataFrame(data)
df.head()
```

```python
df['Type'].value_counts()
```

```python
# excluding the type column and creating a new data
new_df=df.iloc[:,1:]
new_df
```

```python
new_df.shape
```

```python
# info function gives us the raw information of data such as any null values, data types e.t.c.
new_df.info()
```

```python
# Converting data to numpy array
wine_array=new_df.values
wine_array
```

```python
# Normalizing the numerical data 
wine_norm=scale(wine_array)
wine_norm
```

```python
# Applying PCA Fit Transform to dataset
pca=PCA(n_components=13)

wine_pca=pca.fit_transform(wine_norm)
wine_pca
```

```python
# PCA Components matrix or covariance Matrix
pca.components_
```

```python
# obtaining individual variaance values from the ratio
var = pca.explained_variance_ratio_
var
```

```python
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='magenta')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.show()
```

```python
# Cummulative variance of each PCA
variance=np.cumsum(np.round(var,4)*100)
variance
```

```python
# Variance plot for PCA components obtained 
plt.plot(variance,color='magenta')
```

```python
# Final Dataframe
final_df=pd.concat([df['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
# Visualization of PCAs
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df)
```

```python
# As we have already normalized the data, we can directly create Dendrograms
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(wine_norm,'complete'))

# we can see that the the number of clusters is 6
```

```python
#performing hierarchial clustering using first 3 principal component scores
# Create Clusters (y)
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hc
```

```python
y=pd.DataFrame(hc.fit_predict(wine_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
wine_hc=df.copy()
wine_hc['clustersid']=hc.labels_
wine_hc
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#performing # kmeans clustering using first 3 principal component scores i.e k=3
clusters3=KMeans(3,random_state=30).fit(wine_norm)
clusters3
```

```python
clusters3.labels_
```

```python
# Assign clusters to the data set
wine_km=df.copy()
wine_km['clusters3id']=clusters3.labels_
wine_km
```
