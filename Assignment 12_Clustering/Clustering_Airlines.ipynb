```python
std_scaler = StandardScaler()
scaled_data = std_scaler.fit_transform(df)
scaled_data
```

```python
# Normalizing the data by using the data function

def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# creating a Normalize Data Frame

df_norm = norm_func(data.iloc[:,1:])
df_norm
```

```python
#Create Dendrogram after normalizing the data
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(df_norm,method="average"))

# as per the dendrogram there are 3 clusters formed thereby showing that the optimum amount of clusters is 4
# so we go with 3 clusters#
```

```python
#Create Dendrogram to un normalized data
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(df,method="average"))

# as per the dendrogram there are 2 clusters formed thereby showing that the optimum amount of clusters is 4
# so we go with 2 clusters
```

# KMeans clustering

```python
from sklearn.cluster import KMeans
```

```python
ssc = []
for i in range(1,11):
    km = KMeans(n_clusters=i,random_state=2)
    km.fit(scaled_data)
    ssc.append(km.inertia_)
```

```python
plt.plot(range(1,11),ssc)
plt.title('Elbow Method')
plt.xlabel('no.of clusters')
plt.ylabel('SSE')
plt.show()
```

```python
cluster_kmean = KMeans(n_clusters=4)
```

```python
cluster_kmean.fit(scaled_data)
```

```python
cluster_kmean.labels_
```

```python
cluster_kmean = cluster_kmean.labels_
```

```python
 # Assign clusters to the data set
airline_data=df.copy()
airline_data['cluster']=cluster_kmean
airline_data
```

# Hierarchical clustering

```python
x = df.iloc[:,1:]
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler=StandardScaler().fit_transform(x)
```

```python
scaled_data=pd.DataFrame(std_scaler,columns=x.columns)
scaled_data
```

```python
import scipy.cluster.hierarchy as sch
dendogram=sch.dendrogram(sch.linkage(scaled_data,method='single', metric='euclidean'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc=AgglomerativeClustering(n_clusters=2,linkage='single')
y_hc=hc.fit_predict(scaled_data)
y_hc
```

```python
cluster_ag=pd.DataFrame({'cluster':y_hc})
cluster_ag
```

```python
 # Assign clusters to the data set
airline_data1=df.copy()
airline_data1['cluster']=cluster_ag
airline_data1
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
```
