```python
# importing all the required packages and naming them as per our convinience
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# reading the data using pandas
cars=pd.read_csv("E:\Assignment 1/Cars.csv")
cars
```

```python
cars.plot.box()

# the below graph shows us that the data consists of outliers in it and doesnt follow symmetrical distribution
```

```python
cars.plot.hist()
```

```python
sns.distplot(cars.MPG, label='Cars-MPG')
plt.xlabel('MPG')
plt.ylabel('Density')
plt.legend();
```

```python
cars.MPG.mean()

# as per the graph  MPG does follow normal distribution and doesnt contain any outliers in it according to the boxplot
# as the shape of the data is in the form of a bell we can say that mpg follows normal distribution
```
