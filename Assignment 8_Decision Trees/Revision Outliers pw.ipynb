```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset using pandas
data = pd.read_csv("F:\\Assignments\\Completed\\decision Trees\\Company_Data.csv")
data.head()
```

```python
# creating a dataframe using pandas.
df=pd.DataFrame(data)
df
```

```python
df.boxplot()
```

```python
# removing the ouliers using remove_outliers function
def remove_outliers(x):
    sorted_x=sorted(x)
    q1,q3=np.percentile(x,[25,75])
    iqr=q3-q1
    Lower_bound=q3-(1.5*iqr)
    Upper_bound=q3+(1.5*iqr)
    Outliers=[i for i in x if i<Lower_bound or i>Upper_bound]
    new_x=[i for i in x if i<=Lower_bound and i>=Upper_bound]
    return new_x,Outliers

new_Price,Outliers=remove_outliers(df["Price"])
```

```python
#print the number of removed outliers and new length of the income column
print("number of outliers removed:",len(Outliers))
print("new length of the Income column:",len(new_Price))
```
